```python
try:
    database = connection.connect(host="localhost", database='retrowheels', user="root", passwd="", use_pure=True)
    query ="""
    SELECT 
    c.customerName, 
    pr.productName, 
    od.quantityOrdered, 
    od.priceEach 
FROM payments p
JOIN customers c ON p.customerNumber = c.customerNumber
JOIN orders o ON o.customerNumber = c.customerNumber 
    AND p.paymentDate = o.orderDate  -- Pastikan hanya pesanan yang sesuai dengan tanggal pembayaran
JOIN orderdetails od ON o.orderNumber = od.orderNumber
JOIN products pr ON od.productCode = pr.productCode
WHERE p.paymentDate = '2003-06-05' 
AND p.amount = 14571.44;
    """
except Exception as e:
    database.close()
    print(str(e)) 
```
